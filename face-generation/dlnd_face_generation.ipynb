#### ImageFolder

To create a dataset given a directory of images, it's recommended that you use PyTorch's [ImageFolder](https://pytorch.org/docs/stable/torchvision/datasets.html#imagefolder) wrapper, with a root directory `processed_celeba_small/` and data transformation passed in.

```python
# necessary imports
import torch
from torchvision import datasets
from torchvision import transforms
```

```python
def get_dataloader(batch_size, image_size, data_dir='processed_celeba_small/'):
    """
    Batch the neural network data using DataLoader
    :param batch_size: The size of each batch; the number of images in a batch
    :param img_size: The square size of the image data (x, y)
    :param data_dir: Directory where image data is located
    :return: DataLoader with batched data
    """
    # define transforms applied to images upon load
    # DON'T NORMALIZE!! Input images are pre-normalized!
    # DON'T APPLY RANDOM RESIZE CROP!! Input images are very tightly pre-cropped.
    train_tfs = transforms.Compose([
        transforms.RandomHorizontalFlip(),
        transforms.Resize(image_size),
        transforms.ColorJitter(0.2, 0.2, 0.2),
        transforms.ToTensor(),
    ])
    train_set = datasets.ImageFolder(data_dir, transform=train_tfs)
    train_loader = torch.utils.data.DataLoader(train_set, batch_size=batch_size, shuffle=True, num_workers=32)
    
    return train_loader
```

# Define the Model

A GAN is comprised of two adversarial networks, a discriminator and a generator.

## Discriminator

Your first task will be to define the discriminator. This is a convolutional classifier like you've built before, only without any maxpooling layers. To deal with this complex data, it's suggested you use a deep network with **normalization**. You are also allowed to create any helper functions that may be useful.

#### Exercise: Complete the Discriminator class
* The inputs to the discriminator are 32x32x3 tensor images
* The output should be a single value that will indicate whether a given image is real or fake


```python
import torch.nn as nn
import torch.nn.functional as F
```

```python
class Discriminator(nn.Module):

    def __init__(self, conv_dim):
        """
        Initialize the Discriminator Module
        :param conv_dim: The depth of the first convolutional layer
        """
        super(Discriminator, self).__init__()        
        self.conv_dim = conv_dim
        
        self.conv_layers = nn.Sequential(
            
            # 32 x 32 x 3
            nn.Conv2d(3, conv_dim, kernel_size=(3, 3), stride=(1, 1), padding=(1, 1), bias=False),
            nn.LeakyReLU(0.2),
            # 32 x 32 x conv_dim
            nn.Conv2d(conv_dim, conv_dim, kernel_size=(4, 4), stride=(2, 2), padding=(1, 1), bias=False),
            nn.LeakyReLU(0.2),
            nn.BatchNorm2d(conv_dim),
            # 16 x 16 x conv_dim          
            nn.Conv2d(conv_dim, 2 * conv_dim, kernel_size=(4, 4), stride=(2, 2), padding=(1, 1), bias=False),
            nn.LeakyReLU(0.2),
            nn.BatchNorm2d(2 * conv_dim),
            # 8 x 8 x 2 * conv_dim            
            nn.Conv2d(2 * conv_dim, 4 * conv_dim, kernel_size=(4, 4), stride=(2, 2), padding=(1, 1), bias=False),            
            nn.LeakyReLU(0.2),
            nn.BatchNorm2d(4 * conv_dim),
            # 4 x 4 x 2 * conv_dim
        )
        
        self.ff_layers = nn.Sequential(
            nn.Linear(in_features=4 * 4 * 4  *conv_dim, out_features=1)
        )
        

    def forward(self, x):
        """
        Forward propagation of the neural network
        :param x: The input to the neural network     
        :return: Discriminator logits; the output of the neural network
        """
        x = self.conv_layers(x)
        x = x.view(-1, 4 * 4 * 4 * self.conv_dim)
        x = self.ff_layers(x)
#         x = torch.sigmoid(x)
        # for passing on to loss function, need to return scores, not probabilities
        return x


"""
DON'T MODIFY ANYTHING IN THIS CELL THAT IS BELOW THIS LINE
"""
tests.test_discriminator(Discriminator)
```

## Generator

The generator should upsample an input and generate a *new* image of the same size as our training data `32x32x3`. This should be mostly transpose convolutional layers with normalization applied to the outputs.

#### Exercise: Complete the Generator class
* The inputs to the generator are vectors of some length `z_size`
* The output should be a image of shape `32x32x3`

```python
class Generator(nn.Module):
    
    def __init__(self, z_size, conv_dim):
        """
        Initialize the Generator Module
        :param z_size: The length of the input latent vector, z
        :param conv_dim: The depth of the inputs to the *last* transpose convolutional layer
        """        
        super(Generator, self).__init__()
        self.conv_dim = conv_dim
        
        # map z to a 4x4 image with conv_dim*4 channels
        self.fc = nn.Linear(z_size, conv_dim*4*4*4)
        
        self.conv_layers = nn.Sequential(
            
            # 4 x 4 x (4*conv_dim)
            nn.ConvTranspose2d(4*conv_dim, 4*conv_dim, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.BatchNorm2d(4*conv_dim),
            # 4 x 4 x (4*conv_dim)
            nn.ConvTranspose2d(4*conv_dim, 2*conv_dim, kernel_size=4, stride=2, padding=1),
            nn.ReLU(),
            nn.BatchNorm2d(2*conv_dim),
            # 8 x 8 x (2*conv_dim)
            nn.ConvTranspose2d(2*conv_dim, conv_dim, kernel_size=4, stride=2, padding=1),
            nn.ReLU(),
            nn.BatchNorm2d(conv_dim),
            # 16 x 16 x (conv_dim)
            nn.ConvTranspose2d(conv_dim, 3, kernel_size=4, stride=2, padding=1),
            # 32 x 32 x 3
        )
                

    def forward(self, x):
        """
        Forward propagation of the neural network
        :param x: The input to the neural network     
        :return: A 32x32x3 Tensor image as output
        """
        # define feedforward behavior
        x = self.fc(x)
        x = x.view(-1, self.conv_dim*4, 4, 4)
        x = self.conv_layers(x)
        x = 2 * torch.sigmoid(x) - 1
        return x

"""
DON'T MODIFY ANYTHING IN THIS CELL THAT IS BELOW THIS LINE
"""
tests.test_generator(Generator)
```

## Discriminator and Generator Losses

Now we need to calculate the losses for both types of adversarial networks.

### Discriminator Losses

> * For the discriminator, the total loss is the sum of the losses for real and fake images, `d_loss = d_real_loss + d_fake_loss`. 
* Remember that we want the discriminator to output 1 for real images and 0 for fake images, so we need to set up the losses to reflect that.


### Generator Loss

The generator loss will look similar only with flipped labels. The generator's goal is to get the discriminator to *think* its generated images are *real*.

#### Exercise: Complete real and fake loss functions

**You may choose to use either cross entropy or a least squares error loss to complete the following `real_loss` and `fake_loss` functions.**

```python
#
# I was eventually able to get reasonable results using a BCE loss function, but the quality
# was very poor so I decided to also implement the Wasserstein GAN cost function, which is 
# thought to have generally better convergence properties, in particular on the CelebA dataset: 
#    https://towardsdatascience.com/gan-ways-to-improve-gan-performance-acf37f9f59b
# In the end, I don't think it made much of a difference, but I kept the code around anyway.
#
def WGAN_loss(D_out):
    '''
    Calculates how close discriminator outputs are to being real.
    param, D_out: discriminator logits
    return: real loss
    '''
    loss = torch.mean(D_out)
    return loss

def real_loss(D_out):
    '''
    Calculates how close discriminator outputs are to being real.
    param, D_out: discriminator logits
    return: real loss
    '''
    target = torch.ones(D_out.shape[0])
    if train_on_gpu:
        target = target.cuda()
    loss = nn.BCEWithLogitsLoss()(D_out.squeeze(), target)
    return loss

def fake_loss(D_out):
    '''
    Calculates how close discriminator outputs are to being real.
    param, D_out: discriminator logits
    return: real loss
    '''
    target = torch.zeros(D_out.shape[0])
    if train_on_gpu:
        target = target.cuda()
    loss = nn.BCEWithLogitsLoss()(D_out.squeeze(), target)
    return loss

def get_loss(fake_images, real_images=None, use_wgan=False):
    '''
    Get the loss for generator / discriminator, choosing between WGAN loss 
    functions and DCGAN loss formulation.
    '''
    # discriminator
    if real_images is not None:
        if use_wgan: # choose between WGAN and DCGAN loss functions
            loss = WGAN_loss(D(real_images)) - WGAN_loss(D(fake_images))
        else:
            loss = real_loss(D(real_images)) + fake_loss(D(fake_images))
                
    # generator
    else:
        if use_wgan: # choose between WGAN and DCGAN loss functions
            loss = WGAN_loss(D(fake_images))
        else:
            loss = real_loss(D(fake_images))
    
    return loss
```

```python
# helper function for viewing a list of passed in sample images
def view_samples(epoch, samples, display_now=False):
    fig, axes = plt.subplots(figsize=(16,4), nrows=2, ncols=8, sharey=True, sharex=True)
    for ax, img in zip(axes.flatten(), samples[epoch]):
        img = img.detach().cpu().numpy()
        img = np.transpose(img, (1, 2, 0))
        img = ((img + 1)*255 / (2)).astype(np.uint8)
        ax.xaxis.set_visible(False)
        ax.yaxis.set_visible(False)
        im = ax.imshow(img.reshape((32,32,3)))
    if display_now:
        display(fig)
```

```python
def train(D, G, n_epochs, print_every=3000, use_wgan=True):
    '''Trains adversarial networks for some number of epochs
       param, D: the discriminator network
       param, G: the generator network
       param, n_epochs: number of epochs to train for
       param, print_every: when to print and record the models' losses
       return: D and G losses'''
    
    # move models to GPU
    if train_on_gpu:
        D.cuda()
        G.cuda()

    # keep track of loss and generated, "fake" samples
    samples = []
    losses = []

    # Get some fixed data for sampling. These are images that are held
    # constant throughout training, and allow us to inspect the model's performance
    sample_size=32
    fixed_z = np.random.uniform(-1, 1, size=(sample_size, z_size))
    fixed_z = torch.from_numpy(fixed_z).float()
    # move z to GPU if available
    if train_on_gpu:
        fixed_z = fixed_z.cuda()

    # epoch training loop
    for epoch in range(n_epochs):

        # batch training loop
        for batch_i, (real_images, _) in enumerate(celeba_train_loader):

            batch_size = real_images.size(0)
            real_images = scale(real_images).cuda()

            # ===============================================
            #         YOUR CODE HERE: TRAIN THE NETWORKS
            # ===============================================
            
            # 1. Train the discriminator on real and fake images
            z = torch.Tensor(np.random.uniform(-1, 1, size=(batch_size, z_size)))
            if train_on_gpu:
                z = z.cuda()
            fake_images = G(z)
            d_optimizer.zero_grad()
            d_loss = get_loss(fake_images, real_images, use_wgan=use_wgan)
            d_loss.backward()
            d_optimizer.step()

            if use_wgan:
                for p in D.parameters():
                    p.data.clamp_(-0.02, 0.02)

            # 2. Train the generator with an adversarial loss
            z = torch.Tensor(np.random.uniform(-1, 1, size=(batch_size, z_size)))
            if train_on_gpu:
                z = z.cuda()
            fake_images = G(z)
            g_optimizer.zero_grad()
            g_loss = get_loss(fake_images, use_wgan=use_wgan)
            g_loss.backward()
            g_optimizer.step()
            
            # ===============================================
            #              END OF YOUR CODE
            # ===============================================

            # Print some loss stats
            if batch_i % print_every == 0:
                # append discriminator loss and generator loss
                losses.append((d_loss.item(), g_loss.item()))
                # print discriminator and generator loss
                print('Epoch [{:5d}/{:5d}] | d_loss: {:6.8f} | g_loss: {:6.8f}'.format(
                        epoch+1, n_epochs, d_loss.item(), g_loss.item()))

        ## AFTER EACH EPOCH##    
        # this code assumes your generator is named G, feel free to change the name
        # generate and save sample, fake images
        G.eval() # for generating samples
        samples_z = G(fixed_z)
        samples.append(samples_z)
        G.train() # back to training mode

        # check status visually
        if not epoch % 5:
            view_samples(-1, samples, display_now=True)
    
    # Save training generator samples
    with open('train_samples.pkl', 'wb') as f:
        pkl.dump(samples, f)


    # finally return losses
    return losses
```
